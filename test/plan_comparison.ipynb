```python
import numpy
import matplotlib.pyplot as plt

def cost_curve(x,premium,contribution,deductible,coinsuranceRate,maxOutOfPocket):
    eff_premium = premium - contribution
    post_deduct = (coinsuranceRate * (x - deductible)) + deductible
    y = numpy.fmin(numpy.fmin(x, post_deduct), maxOutOfPocket) + eff_premium
    return y
```

```python
x = numpy.linspace(0.0, 10000, 100)
y1 = cost_curve(x,618.96, 300, 3000, 0.65, 7000)
y2 = cost_curve(x,1200.04, 400, 2250, 0.65, 4700)

fig, ax = plt.subplots()
ax.set_xticks([4500,9000],minor=True)
ax.xaxis.grid(True,which='minor')
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('Out-of-Pocket Expenses')
plt.ylabel('Actual Incurred Cost')
plt.legend(['HSA', 'HRA'])
plt.show()
```

```python
print('For expenses = $4,500:')
print('HSA will cost: $'+str(cost_curve(4500,618.96, 300, 3000, 0.65, 7000)))
print('HRA will cost: $'+str(cost_curve(4500,1200.04, 400, 2250, 0.65, 4700)))

print('For expenses = $9,000:')
print('HSA will cost: $'+str(cost_curve(9000,618.96, 300, 3000, 0.65, 7000)))
print('HRA will cost: $'+str(cost_curve(9000,1200.04, 400, 2250, 0.65, 4700)))
```
